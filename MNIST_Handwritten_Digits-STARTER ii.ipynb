# Introduction to MNIST Dataset

MNIST is a popular handwritten numeric dataset containing 60,000 handwritten numbers for machine learning model training and 10,000 handwritten numbers for model testing. This dataset is very useful for testing a model or a new algorithm, if the model has been tested on MNIST and achieves high accuracy, the new algorithm is probably correct. If the algorithm doesn't work on MNIST, it won't work at all.

The following code shows how MNIST dataset is been loaded from `torchvision.datasets`.

```python
# Define transforms
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=(0.1307,), std=(0.3081)),
                                transforms.Lambda(lambda x: torch.flatten(x))])

# Create training set and define training dataloader
training_data = datasets.MNIST(root="data", train=True, download=True, transform=transform)
test_data = datasets.MNIST(root="data", train=False, download=True, transform=transform)

# Create test set and define test loader 
train_loader = DataLoader(training_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size= len(test_data))
```

From the preceeding code, we put three transformations together, first, we converted the NumPy array to tensor and then normalized the tensors, the PyTorch implementation of "normalize" subtracts from each MNIST image the mean 0.1307, for each channel separately and since we have a grayscale image which is 2-D, we have (0.1307), (0.1307), for the first and second channel respectively and then divide by the standard deviation 0.3081. These 0.1307 and 0.3081 are the mean and standard deviation of the training data. This transformation causes each of the input images of our MNIST dataset to be transformed from the pixel value of [0, 255] to [-0.4242129623889923, 2.821486711502075], this will allow our model to converge in the correct direction and also helps to achieve better accuracy of the model, Next we flatten these images before sending them to our model , this process converts the 28X28 dimensional arrays into a 1-D array (1,28*28), we do this because 1-D arrays take less amount of memory compared to multi- Dimension arrays, and thus reduce the time to train the model. Then we specify the root directory called "data" where we want to save the MNIST dataset after download, for training_data we specify that we want to download only the training data by specifying train = True, then we have the transformation applied, for test_data we specify train as False to specify test data. So far we have built a neural network, next we run the network,  We start by creating a function. `batch_training_loss`in the cell below. The function start by obtaining the prediction of our model for every batch in our training loader, then compute how far we are to the actual labels, which is the loss value, we set the gradient to zero, before updating the weights and biases, this according to PyTorch documentation is because PyTorch's computation of gradients is in buffer, since we will be computing the gradients for every 32 batches in our training data, we set the previous grad to zero before computing the next one, so that we can have our weights and biases updated correctly. Next we compute the gadient of the loss and make the optimizer iterates over all the  paramters and use the the stored gradient to update, this function return the updated batch loss.

```python

def batch_training_loss(inputs, labels, model, loss_fn, opt):
    
    # Put the model in the training mode
    model.train()
    
    # predict each training inputs
    outputs = model(inputs)
    
    # compute the error
    batch_loss = loss_fn(outputs, labels)
    
    # zeroing out the grad of the previous batch
    opt.zero_grad()
    
    # compute the gradients of the batch loss 
    batch_loss.backward()
    
    # make the optimizer iterates over all the parameters and use their internally stored gradient to update their values 
    opt.step()
    
    return batch_loss.item()
```

Next, the function `validation_loss` compute the loss value on the validation loader, the need for this function is because the loss of the training set is computed during training process, unlike the `batch_training_loss`, we don't need the gradient and therefore we deactivated it, and this reduce the computational cost and the use of memory.

```python
@torch.no_grad()
def validation_loss(inputs, labels, model,loss_fn):
    
    # Put the model in evaluation mode
    model.eval()
    
    # Compute the prediction of the validation set
    outputs = model(inputs)
    
    # Compute the loss values of the prediction on the validation set
    val_loss = loss_fn(outputs, labels)
    
    return val_loss.item()
```

Next, the funtion `accuracy` compute the prediction of the inputs on our neural network, and return the list of the correct labels.

```python
def accuracy(inputs, labels, model):
    
    # deactivate the autograd to compute accuracy  
    with torch.no_grad():
        
        # put the model in evaluation mode
        model.eval()
        
        # Make the predictions
        output = model(inputs)
        
    #Returns the maximum value and the argument maxes of all elements in our prediction output
    _, argmaxes = output.max(-1)
    
    # check if the predictions are correct with the labels and store as corect.
    correct = argmaxes == labels
    
    return correct.cpu().numpy().tolist()
```

Finally we train the neural network and return the losses and accuracy for each epoch on the training set and the validation set.  

```python
def train_model(num_epch, model, loss_fn, optim, train_dl, test_dl):

    # Set the over all  training loss and over all  training accuracy as empty  
    training_losses, training_accuracies  =  [], []

    # Set the overall validation loss and  validation accuracy as empty 
    validation_losses, validation_accuracies = [], []
    
    
    for epoch in range(num_epch):
        print(epoch)
        
        # Create empty array to save the training losses and training accuracies of each batch as the model train 
        batch_losses, batch_accuracies = [], []
     
        #iterate through the batches of the training data loader which is the variable "data", note that, 
        #the 'data' variable contains one batch of the training loader, we compute the accuracies and losses 
        # enumerate enable us to be able to print status during each epoch expecially when the data is big.    
        
        for i, data in enumerate(iter(train_loader)):
            
            # Get the training inputs and labels of one batch
            inputs, labels = data
            inputs = inputs.to(device)
            labels = labels.to(device)
        
            # Compute the batch loss i.e the loss value with in one batch of the training data
            data_loss = batch_training_loss(inputs, labels, model, loss_fn, optim)
        
            # Append the loss of each batch to the batches_losses variable
            batch_losses.append(data_loss)
        
        # After iterating over all batches, computed each batch loss values in the train loader, 
        #we Compute the average of this batch loses for every epoch(i.e the number times that the 
        #learning algorithms work through the entire training dataset) and this make epoch_training_loss variable
        epoch_training_loss = np.array(batch_losses).mean()
    
        # The overall training loss is the loss of each epoch
        training_losses.append(epoch_training_loss)
    
    
        # Iterate through the batches of the train data loader and compute the  training accuracy
        for i, data in enumerate(iter(train_loader)):
            
            # Get the training inputs and labels of one batch
            inputs, labels = data
            inputs = inputs.to(device)
            labels = labels.to(device)
        
            # Compute the batch accuracy i.e the accuracy value with in one batch of the training data
            correct = accuracy(inputs, labels, model)
        
            # Append the loss of each batch to the batches_losses variable
            batch_accuracies.extend(correct)
        
        # After iterating over all batches, computed each batch accuracies values in the train loader, 
        #we Compute the average of this batch Accuracies  for every epoch and this make "epoch_training_accuracy" variable
        epoch_training_accuracy = np.mean(batch_accuracies)
    
        # each epoch training is the training accuracy
        training_accuracies.append(epoch_training_accuracy)
    
    
    
    
        # Iterate through the batches of the test/validation data loader and compute the validation loss and accuracy
        for i, data in enumerate(iter(test_loader)):
            
            # Get the test inputs and labels and since we are using all the validation as our batch size
            inputs, labels = data
            inputs = inputs.to(device)
            labels = labels.to(device)
        
            # Compute the accuracy of the prediction on the validation set
            val_correct = accuracy(inputs, labels, model)
        
        
            # Compute the loss of the prediction on the validation set
            val_loss = validation_loss(inputs, labels, model, loss_fn)
            
        # E
        epoch_validation_accuracy = np.mean(val_correct)
        validation_accuracies.append(epoch_validation_accuracy)
    
        #
        validation_losses.append(val_loss)
    
    return training_losses, training_accuracies, validation_losses, validation_accuracies       
```

```python
class Net_9(nn.Module):
    def __init__(self):
        super().__init__()
        self.drop= nn.Dropout(0.5)
        self.fc1 = nn.Linear(784, 100)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(100, 100)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(100, 100)
        self.relu3 = nn.ReLU()
        self.drop= nn.Dropout(0.5)
        self.fc4= nn.Linear(100,10)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        x = self.relu3(x)
        x = self.fc4(x)
        return x

model_9 = Net_9().to(device)

# specify Loss function
loss_fn_9 = nn.CrossEntropyLoss()

optimizer_9 = SGD(params= model_9.parameters(),lr=1e-2)
```

```python

def batch_training_loss(inputs, labels, model, loss_fn, opt):
    
    # Put the model in the training mode
    model.train()
    
    # predict each training inputs
    outputs = model(inputs)
    
    #regularization
    l2 = 0
    for param in model.parameters():
        l2 += torch.norm(param,2)
    
    # compute the error
    batch_loss = loss_fn(outputs, labels) + 0.01 * l2
    
    # zeroing out the grad of the previous batch
    opt.zero_grad()
    
    # compute the gradients of the batch loss 
    batch_loss.backward()
    
    # make the optimizer iterates over all the parameters and use their internally stored gradient to update their values 
    opt.step()
    
    return batch_loss.item()
```
